```python
!pip install yfinance
import yfinance as yf
import pandas as pd

# Mapping of coin names to yfinance ticker symbols
coin_tickers = {
    'aave': 'AAVE-USD',
    'binancecoin': 'BNB-USD', 
    'cosmos': 'ATOM-USD',
    'dogecoin': 'DOGE-USD',
    'ethereum': 'ETH-USD',
    'litecoin': 'LTC-USD',
    'nem': 'XEM-USD',
    'stellar': 'XLM-USD',
    'uniswap': 'UNI-USD',
    'wrapped-bitcoin': 'WBTC-USD'
   # Add your selection here
}

# Date range - Specify as desired
start_date = "2021-07-21"
end_date = "2024-06-12"

# Output folder path in Fabric Lakehouse
output_base = "abfss://Group1_CyrptoIntelligence@onelake.dfs.fabric.microsoft.com/Group1CryptoProjectLakehouse.Lakehouse/Files/yfinancehistoricaldata" # Adjust to your path

# Loop through each coin and download + save
for coin_name, ticker in coin_tickers.items():
    print(f"📥 Downloading data for {coin_name} ({ticker})...")
    df = yf.download(ticker, start=start_date, end=end_date, interval="1d")

    if not df.empty:
        df.reset_index(inplace=True)
        df.insert(0, "Symbol", ticker)
        filename = f"{output_base}{ticker}.csv"
        df.to_csv(filename, index=False)
        print(f"✅ Saved to {filename}")
    else:
        print(f"⚠️ No data for {coin_name} ({ticker})")

print("\n🎉 All downloads complete.")
```
